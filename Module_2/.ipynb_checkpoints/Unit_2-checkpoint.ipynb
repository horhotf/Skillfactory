```python
def inspection(stud_math, col):
    #функция осмотра данных
    #Проверка на строки 
    if stud_math.loc[:, col].dtypes == np.dtype("O"):
        fig, ax = plt.subplots(figsize=(6,5))
        sns.countplot(stud_math.loc[:, col], ax=ax)
        plt.show()
    else:
        plot_df = pd.DataFrame({'Распределение':stud_math[col]})
        if (find_miss(stud_math, col)):
            low, high = IQR(stud_math, col)
            plot_df["Выбросы"] = stud_math.loc[stud_math.loc[:, col].between(low, high)][col]
            
        fig, ax = plt.subplots(figsize=(6,5))
        plot_df.hist(ax=ax)
        plt.show()
    miss = (1- (stud_math[col].count() / stud_math.shape[0]))
    print("выбросов", miss)
```

```python
def find_miss(stud_math, col):
    # функция проверки на выбросы
    for index, stud in stud_math.iterrows():
        if (col == "school"):
            if (stud[col] == "GP" or stud[col] == "MS"):
                return True
            else:
                return False
        if (col == "sex"):
            if (stud[col] == "F" or stud[col] == "M"):
                return True
            else:
                return False
        if (col == "age"):
            if (stud[col] >= 15 or stud[col] <= 22):
                return True
            else:
                return False
        if (col == "address"):
            if (stud[col] == "U" or stud[col] == "R"):
                return True
            else:
                return False
        if (col == "Pstatus"):
            if (stud[col] == "T" or stud[col] == "A"):
                return True
            else:
                return False
        if (col == "Medu" or col == "Fedu"):
            if (stud[col] >= 0 or stud[col] <= 4):
                return True
            else:
                return False
        if (col == "Fjob" or col == "Mjob"):
            if (stud[col] == "teacher" or stud[col] == "health" or stud[col] == "services" or stud[col] == "at_home" or stud[col] == "other"):
                return True
            else:
                return False
        if (col == "reason"):
            if (stud[col] == "home" or stud[col] == "reputation" or stud[col] == "course" or stud[col] == "other"):
                return True
            else:
                return False
        if (col == "guardian"):
            if (stud[col] == "mother" or stud[col] == "father" or stud[col] == "other"):
                return True
            else:
                return False
        if (col == "traveltime" or col == "studytime"):
            if (stud[col] >= 1 or stud[col] <= 3):
                return True
            else:
                return False
        if (col == "failures"):
            if (stud[col] >= 0 or stud[col] <= 3):
                return True
            else:
                return False
        if (col == "schoolsup" or col == "famsup" or col == "paid" or col == "activities" or col == "nursery" or col == "higher" or col == "internet" or col == "romantic"):
            if (stud[col] == "yes" or stud[col] == "no"):
                return True
            else:
                return False
        if (col == "famrel" or col == "freetime" or col == "goout" or col == "health"):
            if (stud[col] >= 1 or stud[col] <= 5):
                return True
            else:
                return False
```

```python
def IQR(stud_math, col):
    #Функция расчета пороговых значений
    perc25 = stud_math[col].quantile(0.25)
    perc75 = stud_math[col].quantile(0.75)
    IQR = perc75 - perc25
    low = perc25 - 1.5*IQR
    high = perc75 + 1.5*IQR
    return low, high
```

```python
# Генеротор значений с пороговым значением
import random

def replace_nan(stud_math, col):
    if (stud_math[col].dtypes == np.dtype('O')):
        prob = stud_math[col].value_counts(normalize = True)
        ind = stud_math[stud_math[col].isna()==True].index
        fill_list = pd.Series(random.choices(prob.index, prob, k=stud_math[col].isna().sum()))
        fill_list.index = ind
        return stud_math[col].fillna(fill_list)
    else:
        median = stud_math[col].median()
        return stud_math[col].fillna(median)
    
```

```python
#Формирование нового df

for col in stud_math.columns:
    stud_math[col] = replace_nan(stud_math, col)

stud_math['age_corr'] = stud_math.age.apply(lambda x: 19 if x>18 else x)
stud_math['support'] = np.where((stud_math.schoolsup == "yes") | (stud_math.famsup == "yes") | (stud_math.paid == "yes") | (stud_math.activities == "yes"), "yes", "no")
stud_math['traveltime_corr'] = stud_math.traveltime.apply(lambda x: 2 if x!=1 else 1)
stud_math['famrel_corr'] = stud_math.famrel.apply(lambda x: 2 if x>3 else 1)
stud_math['health_corr'] = stud_math.health.apply(lambda x: 2 if x>3 else 1)
stud_math['absences_corr'] = stud_math.absences.apply(lambda x: 30 if x>29 else x)
stud_math['Pedu'] = stud_math.Fedu + stud_math.Medu
```

```python
def get_stat_dif(column):
    cols = stud_math_corr_cut.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(stud_math_corr_cut.loc[stud_math_corr_cut.loc[:, column] == comb[0], 'score'], 
                        stud_math_corr_cut.loc[stud_math_corr_cut.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
for col in stud_math_corr_cut:
    get_stat_dif(col)
```
