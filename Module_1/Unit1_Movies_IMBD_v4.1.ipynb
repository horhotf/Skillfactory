# 13. У какого режиссера самые большие суммарные кассовые сборы?

# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
answers['24'] = '+Four By Two Productions'
```

```python
import itertools
list_production_companies = list(itertools.chain(*data.production_companies.apply(lambda x: x.split("|")))) # объединение списков в единый
production_companies_statistics = pd.Series(list_production_companies).value_counts() # подсчет уникальных значений

len_title_statistics = {} # массив длин названий фильмов 
for i in range(len(data.original_title)):
    length = len(data.original_title[i])
    mass_company = data.production_companies[i].split("|") # получаем список компаний фильма
    for company in mass_company:
        if company in len_title_statistics: # если знаем такую компанию прибавляем длину названия ее фильма
            len_title_statistics[company] += length
        else: # иначе - добавляем новый со значением длины названия фильма
            len_title_statistics[company] = length


norm_len_title_by_comp = {}
for comp in production_companies_statistics.keys():
    norm_len_title_by_comp[comp] = len_title_statistics[comp] / production_companies_statistics[comp] # сумма сборов режисера
dict(sorted(norm_len_title_by_comp.items(), key=lambda item: item[1], reverse=True)) # сортировка



```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers['27'] = '+Daniel Radcliffe,Rupert Grint'
```

```python
import itertools
list_mass_casts = data.cast.apply(lambda x: x.split("|"))

couple_cast = {}

for film in list_mass_casts:
    for act1 in range(0,len(film)):     
        for act2 in range(act1+1,len(film)):
            string = str(film[act1]) + "," + str(film[act2])
            if string in couple_cast: # если знаем такой прибыльный жанр увеличиваем счетчик
                couple_cast[string] += 1
            else: # иначе - добавляем новый со значением 1
                couple_cast[string] = 1
dict(sorted(couple_cast.items(), key=lambda item: item[1], reverse=True)) # сортировка
```
